### Task 1

```python
train_data = pd.read_csv("../../DataSets/house_price_classification.csv")
train_data.head()
```

```python
train_data.shape
```

```python
train_data.drop(["PriceCategory", "SalePrice"], axis=1)
train_data = train_data.drop(["PoolQC", "MiscFeature"], axis=1)
```

```python
train_cols = train_data.columns

missing_train_cols = []
missing_train_val = []
for col in train_cols:
    if train_data[col].isnull().sum() > 0:
        missing_train_cols.append(col)
        missing_train_val.append(train_data[col].isnull().sum())
        print(f"column {col} have {train_data[col].isnull().sum()} missing values.")
        
```

```python
train_data["LotFrontage"] = train_data["LotFrontage"].fillna(train_data["LotFrontage"].median(), inplace=True)
train_data["MasVnrArea"] = train_data["MasVnrArea"].fillna(train_data["MasVnrArea"].median(), inplace=True)
train_data["GarageYrBlt"] = train_data["GarageYrBlt"].fillna(train_data["GarageYrBlt"].median(), inplace=True)

train_data["Electrical"] = train_data["Electrical"].fillna(train_data["Electrical"].mode()[0], inplace=True)

bsmt_cols = ["BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2"]
for col in bsmt_cols:
    train_data[col] = train_data[col].fillna("None", inplace=True)

garage_num_cols = ["GarageYrBlt"]
garage_cat_cols = ["GarageType", "GarageFinish", "GarageQual", "GarageCond"]

for col in garage_cat_cols:
    train_data[col] = train_data[col].fillna("None", inplace=True)

for col in garage_num_cols:
    train_data[col] = train_data[col].fillna(train_data[col].median(), inplace=True)

high_missing_cols = ["Alley", "FireplaceQu", "Fence"]
for col in high_missing_cols:
    train_data[col] = train_data[col].fillna("None", inplace=True)

train_data["MasVnrType"] = train_data["MasVnrType"].fillna("None", inplace=True)

print(train_data.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# Make sure we’re working with the DataFrame
ohe = OneHotEncoder(sparse_output=False)

# Fit & transform, keep as a separate array
street_encoded = ohe.fit_transform(train_data[["Street"]])

# Convert to DataFrame
street_encoded_df = pd.DataFrame(
    street_encoded, 
    columns=ohe.get_feature_names_out(["Street"]),
    index=train_data.index  # keep the same row indices
)

# Concatenate with original DataFrame (drop old 'Street')
train_data = pd.concat([train_data.drop("Street", axis=1), street_encoded_df], axis=1)
```

### Part A: Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
import pandas as pd
import numpy as np
```

```python
X_train = pd.read_csv("../../DataSets/house_price_classification_clean_train_70.csv")
X_test = pd.read_csv("../../DataSets/house_price_classification_clean_test_30.csv")
```

```python
X_train.shape
```

```python
y_train = X_train["PriceCategory"]
y_test = X_test["PriceCategory"]

X_train = X_train.drop(["SalePrice", "PriceCategory"], axis=1)
X_test = X_test.drop(["SalePrice", "PriceCategory"], axis=1)
```

### Model 1 - Criterion: Entropy, max_dept: 4

```python
dt1 = DecisionTreeClassifier(criterion="entropy", max_depth=4, random_state=42)
dt1.fit(X_train, y_train)
y_pred1 = dt1.predict(X_test)
accuracy1 = accuracy_score(y_test, y_pred1)

print(f"Accuracy of Decision Tree Classifier: {accuracy1:.4f}")
print(classification_report(y_test, y_pred1))
```

### Model 2 - Criterion: Entropy, Max_dept: 5

```python
dt2 = DecisionTreeClassifier(criterion="entropy", max_depth=5, random_state=42)
dt2.fit(X_train, y_train)
y_pred2 = dt2.predict(X_test)
accuracy2 = accuracy_score(y_test, y_pred2)

print(f"Accuracy of Decision Tree Classifier with max_depth=5: {accuracy2:.4f}")
print(classification_report(y_test, y_pred2))
```

### Model 3 --- Criterion: Gini, max_dept: 4

```python
dt3 = DecisionTreeClassifier(criterion="gini", max_depth=4, random_state=42)
dt3.fit(X_train, y_train)
y_pred3 = dt3.predict(X_test)
accuracy3 = accuracy_score(y_test, y_pred3)

print(f"Accuracy of Decision Tree Classifier with Gini: {accuracy3:.4f}")
print(classification_report(y_test, y_pred3))
```

### Model 3 --- Criterion: Gini, max_dept: 5

```python
dt4 = DecisionTreeClassifier(criterion="gini", max_depth=5, random_state=42)
dt4.fit(X_train, y_train)
y_pred4 = dt4.predict(X_test)
accuracy4 = accuracy_score(y_test, y_pred4)

print(f"Accuracy of Decision Tree Classifier with Gini and max_depth=5: {accuracy4:.4f}")
print(classification_report(y_test, y_pred4))
```

```python
from joblib import dump
```

```python
def decision_tree_classifier(X_train, y_train, X_test, y_test, criterion="entropy", max_depth=4, file_name_suffix=None):
    dt = DecisionTreeClassifier(criterion=criterion, max_depth=max_depth, random_state=42)
    dt.fit(X_train, y_train)
    y_pred = dt.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy of Decision Tree Classifier with '{criterion}' and max_depth={max_depth}: {accuracy:.4f}")
    print(classification_report(y_test, y_pred))

    # Saving the model
    file_name = f"./DT_{criterion}_max_depth_{max_depth}_{file_name_suffix}.joblib"
    dump(dt, file_name)

    return accuracy
```

```python
dt_params = [
    {"criterion": "entropy", "max_depth": 4},
    {"criterion": "entropy", "max_depth": 5},
    {"criterion": "gini", "max_depth": 4},
    {"criterion": "gini", "max_depth": 5}
]

train_files = [
    "../../DataSets/house_price_classification_clean_train_70.csv",
    "../../DataSets/house_price_classification_clean_train_75.csv",
    "../../DataSets/house_price_classification_clean_train_80.csv"
]

test_files = [
    "../../DataSets/house_price_classification_clean_test_30.csv",
    "../../DataSets/house_price_classification_clean_test_25.csv",
    "../../DataSets/house_price_classification_clean_test_20.csv"
]

for train_file, test_file in zip(train_files, test_files):
    file_name_suffix = f"{train_file[-6:-4]}_{test_file[-6:-4]}"

    X_train = pd.read_csv(train_file)
    X_test = pd.read_csv(test_file)

    y_train = X_train["PriceCategory"]
    y_test = X_test["PriceCategory"]

    X_train = X_train.drop(["SalePrice", "PriceCategory"], axis=1)
    X_test = X_test.drop(["SalePrice", "PriceCategory"], axis=1)

    print(f"\n===== Evaluating on train file: {train_file[-6:-4]} and test file: {test_file[-6:-4]} =====\n")

    for params in dt_params:
        criterion = params["criterion"]
        max_depth = params["max_depth"]
        accuracy = decision_tree_classifier(X_train, y_train, X_test, y_test, criterion, max_depth, file_name_suffix)
```
