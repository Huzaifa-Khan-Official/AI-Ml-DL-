```python
import numpy as np
import pandas as pd

df = pd.read_csv("./emp.csv")
df

df2 = df.copy()

df.columns


sorted(df2["job"].unique())


df2["job"].value_counts()


# Label Encoding (Custom Mapping)
cols_nums = {"job": {"ANALYST": 3, "CLERK": 5, "MANAGER": 2, "PRESIDENT": 1, "SALESMAN": 4}}

df2 = df2.replace(cols_nums)

df2.head()

def salCatg(salary):
    if salary >= 3000:
        r = 1
    elif salary >= 1500:
        r = 2
    else:
        r = 3
    return r


# Binning (for Salary)
df2["sal"] = df2["sal"].apply(salCatg)
df2.head()

# Automatic Label Encoding using Pandas (category + cat.codes)
df3 = df[df.columns]
df3.dtypes

df3["job"] = df3["job"].astype("category")
df3.dtypes

sorted(df3["job"].unique())

df3["job1"] = df3["job"].cat.codes
df3.head(len(df3))


df3["comm1"] = df3["comm"].astype("category")
df3.dtypes




df3["comm1"] = df3["comm1"].cat.codes
df3.head()
```

```python
df3["cBE_Code"] = np.where(df3["hiredate"].str.contains("2001"), 1, 0)
```
