### Importing required libraries

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import string
```

### Downloading required NLTK data

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
def pos_tagging_analysis(paragraph):
    # Step 1: Preprocessing - Tokenize the input paragraph into words
    print("Step 1: Tokenizing the paragraph...")
    tokens = word_tokenize(paragraph)
    print(f"Tokens: {tokens}\n")
    
    # Step 2: Feature Extraction - Pass tokens into NLTK PoS tagger
    print("Step 2: Applying PoS tagging...")
    pos_tagged = pos_tag(tokens)
    print(f"PoS Tags: {pos_tagged}\n")
    
    # Step 3: Engineering/Mapping - Map tags to human-readable categories
    # Dictionary mapping NLTK tags to grammatical categories
    tag_mapping = {
        # Nouns
        'NN': 'Noun', 'NNS': 'Noun', 'NNP': 'Proper Noun', 'NNPS': 'Proper Noun',
        # Verbs
        'VB': 'Verb', 'VBD': 'Verb', 'VBG': 'Verb', 'VBN': 'Verb', 'VBP': 'Verb', 'VBZ': 'Verb',
        # Adjectives
        'JJ': 'Adjective', 'JJR': 'Adjective', 'JJS': 'Adjective',
        # Adverbs
        'RB': 'Adverb', 'RBR': 'Adverb', 'RBS': 'Adverb',
        # Pronouns
        'PRP': 'Pronoun', 'PRP$': 'Pronoun',
        # Prepositions
        'IN': 'Preposition',
        # Conjunctions
        'CC': 'Conjunction',
        # Determiners
        'DT': 'Determiner',
        # Interjections
        'UH': 'Interjection',
        # Particles
        'RP': 'Particle',
        # Other
        'MD': 'Modal',
        'TO': 'To',
        'EX': 'Existential',
        'FW': 'Foreign Word',
        'LS': 'List Marker',
        'POS': 'Possessive Ending',
        'SYM': 'Symbol',
        'WDT': 'Wh-determiner',
        'WP': 'Wh-pronoun',
        'WP$': 'Wh-pronoun',
        'WRB': 'Wh-adverb'
    }
    
    print("Step 3: Mapping tags to human-readable categories...")
    mapped_tags = []
    for word, tag in pos_tagged:
        category = tag_mapping.get(tag, 'Other')
        mapped_tags.append((word, category))
    
    # Step 4: Output - Display word alongside its grammatical category
    print("\n" + "="*50)
    print("Step 4: Final Output - Words with Grammatical Categories")
    print("="*50)
    for word, category in mapped_tags:
        print(f"{word:15} -> {category}")
    
    return mapped_tags
```

```python
# Test the PoS tagging function
sample_paragraph = "The quick brown fox jumps over the lazy dog near the beautiful river."
print(f"Input Paragraph: {sample_paragraph}\n")
results = pos_tagging_analysis(sample_paragraph)
```
