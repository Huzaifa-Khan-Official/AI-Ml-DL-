### Step 1 — Installing & Importing Dependencies

```python
import re 
import string
import nltk
```

### Downloading required NLTK resources

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize   # Splits text into individual word tokens
from nltk.corpus   import stopwords
```

### Step 2-A — Data Loading (Raw Text)

```python
raw_text = """
Natural Language Processing (NLP) is a fascinating field of Artificial Intelligence!
It enables machines to understand, interpret, and generate Human Language.
Visit https://www.nltk.org for more info. NLTK provides tools for tokenization,
stemming, lemmatization & much more... Don't forget to check out the docs at
http://example.com/docs?ref=nlp#getting-started.
In 2024, NLP models achieved state-of-the-art results on 100+ benchmarks.
<br> Some   extra   whitespace   and HTML tags can also appear in real data. </br>
"""

print('=== RAW TEXT ===')
print(raw_text)
```

### Step 2-B — EDA (Explore Patterns & Encoding)

```python
print(f'Total characters : {len(raw_text)}')
print(f'Total lines      : {len(raw_text.splitlines())}')
```

```python
urls_found = re.findall(r'https?://\S+', raw_text)
print(f'\nURLs detected ({len(urls_found)}):')
for url in urls_found:
    print(f'  {url}')
```

```python
html_tags = re.findall(r'<[^>]+>', raw_text)
print(f'\nHTML tags detected ({len(html_tags)}): {html_tags}')
```

```python
non_ascii = [ch for ch in raw_text if ord(ch) > 127]
print(f'\nNon-ASCII chars: {non_ascii if non_ascii else "None – encoding looks clean"}')
```

### Step 3-A — Noise Removal (URLs, HTML, Whitespace)

```python
text = raw_text

# 1. Strip HTML tags (e.g. <br>, </br>)
text = re.sub(r'<[^>]+>', ' ', text)

# 2. Remove URLs(http / https)
text = re.sub(r'https?://\S+', ' ', text)

# 3. Collapse multiple whitespace / newlines into a single space
text = re.sub(r'\s+', ' ', text).strip()

print('=== AFTER NOISE REMOVAL ===')
print(text)
```

### Step 3-B — Normalization: Lowercase

```python
text_lower = text.lower()

print('=== AFTER LOWERCASING ===')
print(text_lower)
```

### Step 3-C — Tokenization

```python
tokens = word_tokenize(text_lower)

print(f'=== TOKENS (first 30 of {len(tokens)}) ===')
print(tokens[:30])
```

### Step 3-D — Remove Punctuation & Non-Alphabetic Tokens

```python
punct_set = set(string.punctuation)       # Reference set (optional visual check)
print(f'Punctuation chars in scope: {punct_set}\n')
```

```python
tokens_clean = [token for token in tokens if token.isalpha()]

print(f'Tokens before cleaning : {len(tokens)}')
print(f'Tokens after  cleaning : {len(tokens_clean)}')
print(f'\nCleaned tokens (first 30): {tokens_clean[:30]}')
```

### Step 4 — Stopword Removal (Feature Engineering)

```python
stop_words = set(stopwords.words('english'))

print(f'Total English stopwords : {len(stop_words)}')
print(f'Sample: {sorted(list(stop_words))[:20]}')
```

```python
# ── Filter out stopwords ────
tokens_final = [token for token in tokens_clean if token not in stop_words]

print(f'\nTokens before stopword removal : {len(tokens_clean)}')
print(f'Tokens after  stopword removal : {len(tokens_final)}')
print(f'Tokens removed                 : {len(tokens_clean) - len(tokens_final)}')
```

### Step 5-A — Final Output

```python
print('=== FINAL CLEAN TOKENS ===')
print(tokens_final)
print(f'\nTotal meaningful tokens : {len(tokens_final)}')
```

### Step 5-B — Frequency Distribution

```python
from nltk.probability import FreqDist

freq_dist = FreqDist(tokens_final)

print('=== TOP 10 MOST FREQUENT TOKENS ===')
for word, freq in freq_dist.most_common(10):
    print(f'  {word:<20} -> {freq}')
```

### Step 5-C — Reusable clean_text() Function

```python
def clean_text(raw: str, language: str = 'english') -> list:
    """
    Full text-cleaning pipeline.

    Parameters
    ----------
    raw      : str  – Raw input text (URLs, HTML, mixed case, etc.)
    language : str  – NLTK stopword language (default: 'english')

    Returns
    -------
    list – Clean, lowercase, alphabetic tokens with stopwords removed.
    """
    text = re.sub(r'<[^>]+>',  ' ', raw) # 1. Remove HTML tags
    text = re.sub(r'https?://\S+', ' ', text) # 2. Remove URLs
    text = re.sub(r'\s+', ' ', text).strip() # 3. Collapse whitespace
    text = text.lower() # 4. Lowercase
    tokens = word_tokenize(text) # 5. Tokenize
    tokens = [t for t in tokens if t.isalpha()] # 6. Alpha-only filter
    sw = set(stopwords.words(language))
    tokens = [t for t in tokens if t not in sw] # 7. Stopword removal
    return tokens
```

```python
# ── Smoke test ──
result = clean_text(raw_text)
print('Output:', result)
print(f'\nReady for vectorization – {len(result)} tokens.')
```
