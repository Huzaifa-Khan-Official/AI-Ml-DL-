### Importing necessary Libraries

```python
import numpy as np
```

```python
logits = np.array([0.2, 1.5, 0.8, 0.3, 2.1, 0.4, 
                   0.6, 0.9, 1.2, 0.7, 1.8, 0.5])

print("Raw logits (12‑element vector):")
print(logits)
print()
```

### Step 2: Predefined target vectors for classes 'X' and 'O'

```python
target_vector_X = np.array([2.0, 1.5, 1.0, 1.8, 2.2, 0.5, 
                             0.3, 0.6, 0.9, 1.1, 1.4, 0.8])

target_vector_O = np.array([0.5, 0.8, 1.2, 0.9, 0.6, 1.5,
                             1.8, 2.0, 1.3, 1.0, 0.7, 1.1])
```

### Step 3: Feature extraction – similarity score (Softmax / Cosine / Euclidean)

```python
def softmax_probability(input_vector, target_vector):
    """
    Exponential similarity scaled by temperature.
    Returns a probability-like score (higher = more similar).
    """

    similarity = np.dot(input_vector, target_vector)
    
    return np.exp(similarity)
```

```python
def euclidean_similarity(input_vector, target_vector):
    """Euclidean distance converted to similarity (1 / (1 + distance))."""
    distance = np.linalg.norm(input_vector - target_vector)
    return 1 / (1 + distance)
```

```python
# Compute similarity scores using the Softmax‑inspired method
score_X = softmax_probability(logits, target_vector_X)
score_O = softmax_probability(logits, target_vector_O)

print("Similarity scores (Softmax‑style):")
print(f"Class X: {score_X:.4f}")
print(f"Class O: {score_O:.4f}")
```

```python
# Convert to actual probabilities (proper Softmax)
exp_logits = np.exp([score_X, score_O])
probabilities = exp_logits / np.sum(exp_logits)

print(f"\nActual probabilities: X = {probabilities[0]:.4f}, O = {probabilities[1]:.4f}")
```

### Step 4: Performance/classification

```python
def classify(logits, target_X, target_O):
    """Return 'X' if logits are more similar to target_X, else 'O'."""
    score_X = softmax_probability(logits, target_X)
    score_O = softmax_probability(logits, target_O)
    return 'X' if score_X > score_O else 'O'
```

```python
final_label = classify(logits, target_vector_X, target_vector_O)
print("\n" + "="*40)
print(f"FINAL PREDICTION: Class = {final_label}")
print("="*40)

print("\n--- Testing with a different input ---")
unknown_logits = np.array([1.0] * 12)
print(f"Unknown logits (all 1.0):")
pred = classify(unknown_logits, target_vector_X, target_vector_O)
print(f"Prediction: {pred}")
```
