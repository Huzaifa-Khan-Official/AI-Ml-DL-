### Step 1: Importing Neccessary Dependencies

```python
import numpy as np
```

###  Step 2: Input Preparation

```python
input_map = np.array([
    [ 1,  3,  2,  4,  1,  0],
    [ 5,  6,  1,  2,  3,  2],
    [ 3,  2,  7,  8,  1,  4],
    [ 1,  0,  3,  6,  2,  5],
    [ 4,  5,  1,  2,  9,  3],
    [ 2,  1,  3,  4,  6,  8]
])

print("=== INPUT FEATURE MAP ===")
print(input_map)
print(f"\nInput shape : {input_map.shape}")
```

### Step 3: Pooling Setup

```python
pool_size = 2
stride    = 2
```

```python
input_h, input_w = input_map.shape
output_h = (input_h - pool_size) // stride + 1
output_w = (input_w - pool_size) // stride + 1
```

```python
print(f"Pool size   : {pool_size}×{pool_size}")
print(f"Stride      : {stride}")
print(f"Output size : {output_h}×{output_w}")
```

```python
# Pre-allocating the output matrix with zeros
output_map = np.zeros((output_h, output_w), dtype=input_map.dtype)
```

### Step 4 & 5: Pooling Loop + Reconstruction

```python
for i in range(output_h):
    for j in range(output_w):

        # Computing the top-left corner of the current 2×2 block
        row_start = i * stride
        col_start = j * stride

        # Slicing out the 2×2 block from the input feature map
        block = input_map[row_start : row_start + pool_size,
                          col_start : col_start + pool_size]

        # Appling max pooling: take the single highest value in the block
        output_map[i, j] = np.max(block)

        print(f"Block [{i},{j}]  rows {row_start}:{row_start+pool_size}"
              f"  cols {col_start}:{col_start+pool_size}")
        print(block)
        print(f"  → max = {output_map[i, j]}\n")
```

### Step 6: Final Output

```python
print("=== OUTPUT FEATURE MAP (after 2×2 Max Pooling) ===")
print(output_map)
print(f"\nInput  shape : {input_map.shape}")
print(f"Output shape : {output_map.shape}")
print(f"Size reduced by factor : {input_map.size // output_map.size}×")
```

### Reusable Function

```python
def max_pool2d(feature_map: np.ndarray, pool_size: int = 2, stride: int = 2):
    h, w = feature_map.shape
    out_h = (h - pool_size) // stride + 1
    out_w = (w - pool_size) // stride + 1
    output = np.zeros((out_h, out_w), dtype=feature_map.dtype)

    for i in range(out_h):
        for j in range(out_w):
            r, c = i * stride, j * stride
            output[i, j] = np.max(
                feature_map[r:r + pool_size, c:c + pool_size]
            )
    return output
```

```python
# Smoke test on a random 8×8 map
np.random.seed(42)
test_map = np.random.randint(0, 100, size=(8, 8))
print("Random 8×8 input:\n", test_map)
print("\nPooled output:\n", max_pool2d(test_map))
```
