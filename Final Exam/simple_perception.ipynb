### Step 1: Importing Necessary Libraries

```python
import numpy as np
```

### Step 2: Making a Class

```python
class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=10):
        # Step 1: Initialisation – define weights (w) and bias (b)
        self.weights = np.random.randn(input_size) * 0.01   # small random weights
        self.bias = 0.0
        self.learning_rate = learning_rate
        self.epochs = epochs
        
    def step_function(self, z):
        """Step activation: 1 if z >= 0 else 0."""
        return 1 if z >= 0 else 0
    
    # Step 2: Model training (Forward Pass) – calculate weighted sum
    def predict(self, X):
        """
        Step 2: Weighted sum – calculate z = Σ(x_i * w_i) + b
        Step 3: Activation – apply step function
        """
        z = np.dot(X, self.weights) + self.bias   # weighted sum (perceptron equation)
        return self.step_function(z)               # activation
    
    def train(self, X_train, y_train):
        """Simple training loop using the perceptron learning rule."""
        print(f"Training perceptron for {self.epochs} epochs...")
        for epoch in range(self.epochs):
            errors = 0
            for xi, target in zip(X_train, y_train):
                prediction = self.predict(xi)
                error = target - prediction
                if error != 0:
                    # Update weights and bias
                    self.weights += self.learning_rate * error * xi
                    self.bias += self.learning_rate * error
                    errors += 1
            print(f"Epoch {epoch+1}/{self.epochs} – Misclassifications: {errors}")
        print("Training complete.")
```

### Step 3: Testing

```python
X_train = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y_train = np.array([0, 0, 0, 1])

# Create and train perceptron
p = Perceptron(input_size=2, learning_rate=0.1, epochs=5)
p.train(X_train, y_train)

# Test predictions
print("\nPredictions on training data:")
for xi in X_train:
    pred = p.predict(xi)
    print(f"Input: {xi} → Prediction: {pred}")

# Test a custom input
test_input = np.array([1, 0])
print(f"\nCustom input {test_input} → Predicted class: {p.predict(test_input)}")
```
