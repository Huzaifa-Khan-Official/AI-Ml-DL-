### Stop-Words
Stopwords are words that don't contribute to the meaning of sentence

```python
import nltk # natural language toolkit
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
print(stopwords.words("english"))
```

```python
# remove english stopwords function
def remove_stopwords(text, language):
    stop_words = set(stopwords.words(language))
    text = text.lower()
    word_tokens = text.split()
    filtered_text = [word for word in word_tokens if word not in stop_words]
    print("Language ==>", language)
    print("Filtered text ==>", filtered_text)
```

```python
en_text = "This is a sample sentence and we are going to remove the stopwords from this sentence"
remove_stopwords(en_text, "english")
```

### URLs

```python
def remove_urls(text):
    url_pattern = re.compile(r"https://\S+|www\.\S+")
    return url_pattern.sub(r"", text)
```

```python
text = "I hope it will be a useful article for you. Follow me for more update at the following link: https://abc.com/@def"
remove_urls(text)
```

### HTML tags

```python
text = """<html><div>
<h1>Natural Language Processing</h1>
<p>Text Processing for NLP.</p>
<a href="https://abc.com/@nlp">ABC account</a>
</div></html>
"""

html_tags_pattern = r"<.*?>"

text_without_html_tags = re.sub(html_tags_pattern, "", text)

print(text_without_html_tags)
```
