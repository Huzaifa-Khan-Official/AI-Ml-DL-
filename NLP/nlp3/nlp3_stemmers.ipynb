```python
from nltk.corpus import stopwords
stopwords_ = set(stopwords.words("english"))
```

```python
sample_text = "this is a sample text"
tokens = sample_text.split()
clean_tokens = [t for t in tokens if not t in stopwords_]
clean_text = " ".join(clean_tokens)
print(clean_tokens)
print(clean_text)
```

### PorterStemmer

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
words = ["running", "files", "happily", "better", "jumps", "happiest"]
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

### Lancter Stemmer

```python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("english")
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

### RegExp Stemmer

```python
from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer("ing$es$|ly$|er$", min=4)
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

### WordNet Lemmatizer

```python
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
print(lemmatized_words)
```

### TextBlob Lemmatizer

```python
from textblob import Word
```

```python
lemmatized_words = [Word(word).lemmatize() for word in words]
print(lemmatized_words)
```

```python
sentence = "The quick brown fox jumps over the lazy dog"
words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)

print("")
print("POS Tags\n", pos_tags, "\n")

# Print the description
for (word, tag) in pos_tags:
    print(nltk.help.upenn_tagset(tagpattern=tag))
```

```python
from nltk.tree import Tree
nltk.download('words')
```

```python
sentence = "The big brown dog barked loudly."
words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)

chuncked = nltk.ne_chunk(pos_tags)

req_exp = "NP: {<DT>?<JJ>*<NN>}"
rp = nltk.RegexpParser(req_exp)
result = rp.parse(chuncked)


result
```
