```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
df = pd.read_csv("./CarPrice_Assignment.csv")
```

```python
df.shape
```

```python
df.columns
```

```python
num_col = df.select_dtypes(exclude= "object")
```

```python
num_col.columns
```

```python
colsDrop = ["symboling", 'carheight', 'stroke', 'compressionratio', 'peakrpm', "CarName"]
df.drop(columns=colsDrop, inplace=True)
```

```python
cat_col = df.select_dtypes(include = ["object"]).columns
df[cat_col].nunique()
```

```python
df = pd.get_dummies(columns=['fueltype', 'aspiration', 'doornumber', 'drivewheel', 'enginelocation'], data=df, drop_first=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
for i in df.select_dtypes(include='object').columns:
    df[i] = encoder.fit_transform(df[[i]])
```

```python
X = df.drop('price', axis=1)
y = df["price"]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge
```

### Model 1: Multiple Linear Regrssion (All Features)

```python
rgsr = LinearRegression()
```

```python
model = rgsr.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("-------------MLR (All Features)")
print(f"Mean Absoulute Error: {mse:.2f}")
print(f'Root Mean Squared Error: {rmse:.2f}')
print(f"R-Squared Score = {r2*100:.2f} %\n")
```
