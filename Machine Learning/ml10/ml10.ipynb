### Bar Plot

```python
import seaborn as sns
```

```python
df02.values
```

```python
cols = df01.columns
```

```python
plt.figure(figsize=(15, 6))
plt.xticks(range(len(df02)), cols)
plt.title("Bar Plot")
plt.xlabel("Phone")
plt.ylabel("Sale")

chrt = sns.barplot(x=df02.index, y=df02.values)
for p in chrt.patches:
    chrt.annotate(format(p.get_height(), ".0f"),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha = 'center', va = "center", xytext = (0, 9),
                  textcoords = "offset points"
                  )
```

### Line Graph

```python
plt.figure(figsize=(10, 5))
plt.xticks(range(len(df02)), cols)
plt.title("Line Graph")
plt.xlabel("Phone")
plt.ylabel("Sale")
plt.xticks(rotation=5, ha="right")

for a, b in zip(df02.index, df02.values):
    plt.text(a, b, str(b))

plt.plot(df02, marker="o", c = "blue")
```

### Pi Chart

```python
plt.figure(figsize=(10, 10))
plt.title("Pie Chart")
plt.pie(df02, labels=cols, labeldistance=1.15, autopct="%1.2f%%", wedgeprops={"linewidth":3, "edgecolor": "white"})
```

### Stack Bar

```python
ax = df.plot.bar(stacked=True, figsize=(10, 7), rot=0)
plt.title("Stacked Chart")
plt.xlabel("Quarter")
plt.ylabel("Sale")
plt.xticks(range(len(df)), df.Quarter)

for c in ax.containers:
    ax.bar_label(c, label_type="center")

ax.legend(title="Categories", bbox_to_anchor=(1.05, 1), loc="upper left")
```

### UnStacked Bar Chart

```python
chrt = df01.plot(kind="bar", figsize=(15, 7))

plt.xticks(range(len(df)), df.Quarter)
plt.title("UnStacked Chart")
plt.xlabel("Quarter")
plt.ylabel("Sale")

for p in chrt.patches:
    chrt.annotate(format(p.get_height(), ".0f"),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center', xytext=(0, 1),
                  textcoords='offset points')
```
