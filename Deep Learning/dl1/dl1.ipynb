```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-10, 10, 100)
```

### Step Function

```python
def step(x):
    res = 1 if x >= 0 else 0
    return res
```

```python
y = [step(i) for i in x]
plt.plot(x, y)
plt.show()
```

### Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
y = [sigmoid(i) for i in x]
plt.plot(x, y)
plt.show()
```

### Leaky ReLU

```python
def leakyReLU(x):
    res = x if x > 0 else 0.01*x
    return res
```

```python
y = [leakyReLU(i) for i in x]
plt.plot(x, y)
plt.show()
```

### Tanh Function

```python
def tanh(x):
    return (2 / (1 + np.exp(-2 * x)))
```

```python
y = [tanh(i) for i in x]
plt.plot(x, y)
plt.show()
```

### Softplus

```python
def softPlus(x):
    return np.log(1 + np.exp(x))
```

```python
y = [softPlus(i) for i in x]
plt.plot(x, y)
plt.show()
```

### ReLU Function

```python
def relu(x):
    return x if x > 0 else 0
```

```python
y = [relu(i) for i in x]
plt.plot(x, y)
plt.show()
```

### Softmax Function

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / np.sum(e_x, axis=0)
```

```python
y = softmax(x)
plt.plot(x, y)
plt.show()
```
